```python
import pandas as pd
import plotly
import plotly.express as px
```

```python
DATA_URL = 'https://raw.githubusercontent.com/DISHANTJOSHI1/dishant_data690/main/part_B/data/wdx_long.csv'
```

```python
df = pd.read_csv(DATA_URL)

df.sample(4)
```

```python
df.drop(columns=["Unnamed: 0"], inplace=True)
df.head()
```

```python
df_pivot = df.pivot_table(
    index=["Year","Country Code",	"Country Name", "Region",	"Income Group",	"Lending Type"], 
    columns="indicator", 
    values="value"
)

df_pivot.head()
```

```python
df_pivot.reset_index(inplace=True)

df_pivot.head()
```

```python
df_2019_pop = df.query("Year == 2019").query("indicator == 'NY.GDP.PCAP.PP.CD'")
```

```python
df_2019_pop = df.query("Year == 2019").query("indicator == 'NY.GDP.PCAP.PP.CD'")
df_2019_pop.sample(5)
```

```python
df_2019_pop_sort = df_2019_pop.sort_values(by="value", ascending=False)

df_2019_pop_sort.head()
```

```python
fig = px.bar(
    df_2019_pop_sort, 
    x="Country Name", 
    labels={"value":"2019 GDP per capita (current international $)", "Country Name":""},
    y="value",
    color="Country Name",
    height=600,
    width=1000,
    template=list(plotly.io.templates.keys())[5]
)

fig.update_layout(showlegend=True)

fig.show()
```

Above graph shows GDP per capita. Russia has highest per capita income out of the all selected counties followed by China. Whereas countires like India, Bangladesh and pakistan have low GDP per capita. We will try to examine if this affects health of the population of respective country in any way.

```python
df_2019_pop_sort = df_2019_pop_manf.sort_values(by="value", ascending=False)

df_2019_pop_sort.head()
```

Manufacturing is a huge deal when it comes to development. we just saw through above graph, how manufecturing industries performed in the year 2019 of different countries.

Russia and China which have high GDP also have high death by non communal diseases with an average of ~89 each when compare to other given countries. Inda, Pakistan and South Africa have lowest death by non communicable diseases.

```python
df_gdp = df.query("indicator == 'SH.STA.SUIC.P5'")
fig = px.line(
    df_gdp, 
    x="Year", 
    labels={"value":"Sucide mortality rate (per 100,000 population)"},
    y="value",
    color="Country Name",
    height=800,
    width=1000,
    template=list(plotly.io.templates.keys())[5],
)

fig.update_layout(showlegend=True)
fig.update_xaxes(dtick="Y1")       # tick interval very year.
fig.show()
```

```python
df_gdp = df.query("indicator == 'SP.DYN.LE00.IN'")
fig = px.line(
    df_gdp, 
    x="Year", 
    labels={"value":"Life expectency at birth, total (years)"},
    y="value",
    color="Country Name",
    height=800,
    width=1000,
    template=list(plotly.io.templates.keys())[5],
)

fig.update_layout(showlegend=True)
fig.update_xaxes(dtick="Y1")       # tick interval very year.
fig.show()
```

### Suicide and GDP Per capita Correlation

```python
df_pivot_2009=df_pivot.query("Year == 2009")
df_pivot_2009
```

```python
fig = px.scatter(df_pivot_2009, x="NY.GDP.PCAP.PP.CD", 
                 y="SH.STA.SUIC.P5", 
                 trendline="ols", 
                 labels={"NY.GDP.PCAP.PP.CD":"2009 GDP per capita PPP (Current international $)","SH.STA.SUIC.P5":"2009 sucide mortality rate (per 100,000 Population)"},
                 height=600,
                 width=1000,
                 template=list(plotly.io.templates.keys())[5],
                 title="Suicide and GDP Per capita Correlation")
fig.show()
```

In many countries across the world, suicide has become a major public health concern. Economic development does not guarentee a decrease in suicide rates. In fact more with increase in wealth people are becoming more mentally stressed due to work load they face to make the wealth. From the above graph, which shows suicide mortalty rate vesus GDP per capita, we see there is an upward relation between the two components, which proves money is not an indicator of good mental health.
